```python
import matplotlib.pyplot as plt
```

```python
# distributed age value
sns.set()
plt.figure(figure=(6,6))
sns.distplot(insurance_dataset['age'])
plt.title('Age Distribution')
plt.show()
```

```python
# gender coloumn
plt.figure(figure=(6,6))
sns.countplot(x='sex' , data=insurance_dataset)
plt.title('Sex Distribution')
plt.show()
```

```python
insurance_dataset['sex'].value_counts()
```

```python
sns.set()
plt.figure(figsize=(6, 6))
sns.histplot(insurance_dataset['bmi'], kde=True)
plt.title('BMI Distribution')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(x='children', data=insurance_dataset)
plt.title('children')
plt.show()
```

```python
insurance_dataset['children'].value_counts()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(x='smoker', data=insurance_dataset)
plt.title('smoker')
plt.show()
```

```python
insurance_dataset['smoker'].value_counts()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(x='region', data=insurance_dataset)
plt.title('region')
plt.show()
```

```python
insurance_dataset['region'].value_counts()
```

```python
plt.figure(figsize=(6, 6))  # ✅ Fix: use 'figsize', not 'figure'
sns.histplot(insurance_dataset['charges'], kde=True)  # ✅ Fix: use 'histplot' instead of deprecated 'distplot'
plt.title('Charges Distribution')
plt.show()
```
